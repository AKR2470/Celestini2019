```python
def create(dataset,look_back=1):
    x,y=[],[]
    for i in range(len(dataset)-look_back):
        for wgt,cls in dataset[i:(i+look_back)].values:
              x.append([wgt,cls])
        for wgt,cls in dataset[(i+look_back):(i+look_back)+1].values:
            y.append(wgt)
       
    return np.array(x),np.array(y)

look_back=1
create(Dataset,look_back)
```

```python
training_set = df_train.values
ar1=[]
ar2=[]
for i in range(len(training_set)):
    ar1.append(training_set[i][0])

for i in range(len(training_set)):
    ar2.append(training_set[i][1])    
    
ar1 = np.reshape(ar1,(len(ar1),1))
ar2 = np.reshape(ar2,(len(ar2),1))




test_set = df_test.values
ar3=[]
ar4=[]
for i in range(len(test_set)):
    ar3.append(test_set[i][0])

for i in range(len(test_set)):
    ar4.append(test_set[i][1])    
    
ar3 = np.reshape(ar3,(len(ar3),1))
ar4 = np.reshape(ar4,(len(ar4),1))

from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()

ar1 = scale.fit_transform(ar1)
ar2 = scale.fit_transform(ar2)

ar3 = scale.transform(ar3)
ar4 = scale.transform(ar4)

test_set = pd.DataFrame()
training_set = pd.DataFrame()


#test_set['Weighted_Price'] = ar3
#test_set['Close'] = ar4

#test_set
ar1=ar1.flatten()
ar2=ar2.flatten()
ar3=ar3.flatten()
ar4 = ar4.flatten()
test_set['Weighted_Price'] = ar3
test_set['Close'] = ar4

training_set['Weighted_Price'] = ar1
training_set['Close'] = ar2

training_set
```

```python
look_back=1
X_train,Y_train = create(training_set,look_back)
x_test,y_test = create(test_set,look_back)

X_train = np.reshape(X_train,(len(X_train),1,X_train.shape[1]))
x_test = np.reshape(x_test,(len(x_test),1,x_test.shape[1]))

print(X_train[:5])
print(Y_train[:5])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,LSTM,BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential()
model.add(LSTM(256,input_shape=(X_train.shape[1:]),return_sequences=True))

model.add(LSTM(256,input_shape=(X_train.shape[1:])))




model.add(Dense(1))

model.compile(loss='mean_squared_error',optimizer='adam')

res = model.fit(X_train,Y_train,epochs=100,batch_size=16,shuffle=False,
                validation_data=(x_test,y_test),
                callbacks=[EarlyStopping(monitor='val_loss',min_delta=5e-5,patience=20,verbose=1)])

```
